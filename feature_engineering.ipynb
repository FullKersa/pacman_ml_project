```python
from src import utils
from imblearn.over_sampling import RandomOverSampler

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
deserialize_data = utils.deserialize_data

X_train_prep = deserialize_data('data/processed/X_train_prep.pkl')

y_train_prep = deserialize_data('data/processed/y_train_prep.pkl')
```

```python
ros = RandomOverSampler(random_state=42)
```

```python
df_y_train = pd.DataFrame(y_train_prep, columns=["loan_status"])

# Buat grafik countplot
plt.figure(figsize=(6, 4))
sns.countplot(data=df_y_train, x="loan_status", hue="loan_status")

plt.title("Distribusi Label Sebelum Oversampling")
plt.xlabel("Loan Status")
plt.ylabel("Jumlah")
plt.tight_layout()
plt.show()
```

```python
# Lakukan oversampling pada data
X_train_ros, y_train_ros = ros.fit_resample(X_train_prep, y_train_prep)

df_y_train_ros = pd.DataFrame(y_train_ros, columns=["loan_status"])

# Buat grafik countplot
plt.figure(figsize=(6, 4))
sns.countplot(data=df_y_train_ros, x="loan_status", hue="loan_status")

plt.title("Distribusi Label Setelah Oversampling")
plt.xlabel("Loan Status")
plt.ylabel("Jumlah")
plt.tight_layout()
plt.show()
```

```python
# Serialize X_train_ros dan y_train_ros

serialize_data = utils.serialize_data

serialize_data(X_train_ros, "data/processed/X_train_ros.pkl")
serialize_data(y_train_ros, "data/processed/y_train_ros.pkl")
```
